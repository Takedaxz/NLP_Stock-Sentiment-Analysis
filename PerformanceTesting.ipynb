```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.DataFrame({
    "Date": ["2024-01-02 21:30:00", "2024-01-03 01:30:00", "2024-01-03 21:30:00", "2024-01-04 01:30:00", "2024-01-04 21:30:00", "2024-01-05 01:30:00", "2024-01-05 21:30:00"],
    "Sentiment Score": [0.6, -0.6, -0.8, -0.9, -0.8, -0.7, -0.7],
})
```

```python
from IPython.display import display

display(df.head())
```

```python
import yfinance as yf

ticker = "AAPL"
stock_data = yf.download(ticker, start="2024-01-01", end="2024-01-20",multi_level_index=False,interval="4h")
stock_data
```

```python
import pytz

stock_data = stock_data.reset_index()

thai_timezone = pytz.timezone('Asia/Bangkok')
stock_data['Datetime'] = pd.to_datetime(stock_data['Datetime'])
stock_data['Local_Datetime'] = stock_data['Datetime'].dt.tz_convert(thai_timezone)
```

```python
stock_data
```

```python
df["Signal"] = np.where(df["Sentiment Score"] > 0.5, "BUY", 
                        np.where(df["Sentiment Score"] < -0.5, "SELL", "HOLD"))
```

```python
# Convert the 'Date' column in df to datetime with the same timezone as 'Local_Datetime'
df['Date'] = pd.to_datetime(df['Date']).dt.tz_localize(thai_timezone)

# Merge the dataframes
df = df.merge(stock_data, left_on="Date", right_on="Local_Datetime", how="left")
df.drop(columns=["Local_Datetime","Datetime","High","Low","Open","Volume",], inplace=True)
df
```

```python
df['Close'] = stock_data['Close']
```

```python
df['%Change'] = df['Close'].pct_change()
df['%Change'] = df['%Change'].shift(-1)
df['%Change'] = df['%Change'].fillna(0)
df['%Change'] = df['%Change'].replace([np.inf, -np.inf], 0)
df['%Change'] = df['%Change'].astype(float)
```

```python
df.head()
```

# 1. Correlation Testing

```python
correlation_df = df[['Sentiment Score', '%Change']].copy()
correlation_df.fillna(0, inplace=True)
correlation_df
```

```python
import plotly.express as px
fig3 = px.imshow(correlation_df.corr(), title='Correlation between Sentiment Score and %Change')
fig3.show()
```

# 3.Trading

```python
initial_balance = 100000
capital = initial_balance
position = 0

for i in range(len(df)):
    if df["Signal"][i] == "BUY" and position == 0:
        position = capital / df["Close"][i]
        capital = 0
        print(f"Buying at {df['Close'][i]} on {df['Date'][i]}")
    elif df["Signal"][i] == "SELL" and position > 0:
        capital = position * df["Close"][i]
        position = 0
        print(f"Selling at {df['Close'][i]} on {df['Date'][i]}")
    elif df["Signal"][i] == "HOLD":
        print(f"Holding at {df['Close'][i]} on {df['Date'][i]}")

final_value = capital if capital > 0 else position * df["Close"].iloc[-1]
returns = (final_value - initial_balance) / initial_balance * 100
print(f"\nProfit/Loss: {final_value - initial_balance:.2f}")
print(f"Portfolio Return: {returns:.2f}% from {initial_balance} to {final_value:.2f} over {len(df)} days")
```

# 4.Classification Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

df["Actual Movement"] = df["%Change"].apply(lambda x: 1 if x >= 0 else -1)
df["Predicted Movement"] = df["Sentiment Score"].apply(lambda x: 1 if x >= 0 else -1)
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(df["Actual Movement"], df["Predicted Movement"])
fig2=px.imshow(cm, text_auto=True, labels={'x': 'Predicted', 'y': 'Actual'}, title='Confusion Matrix')
fig2.update_xaxes(side="top")
fig2.update_traces(textfont_size=20)
fig2.update_layout(title_x=0.5, title_y=0.95)
fig2.update_layout(xaxis_title="Predicted Movement", yaxis_title="Actual Movement")
fig2.update_layout(width=600, height=600)
fig2.update_traces(textfont_color='white')

fig2.show()
```

```python
print(classification_report(df["Actual Movement"], df["Predicted Movement"], target_names=['Down', 'Up']))
```
