```python
import re

def clean_text(text):
    text = re.sub(r'<[^>]+>', '', text)  # ‡∏•‡∏ö HTML tags
    text = re.sub(r'[^\w\s]', '', text)  # ‡∏•‡∏ö special characters
    text = text.lower()  # ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô lowercase
    return text
```

```python
import pandas as pd
```

```python
df_1 = pd.read_csv('dataset_1.csv', encoding='ISO-8859-1', header=None)
df_1
```

```python
df_1.columns = ['label', 'text']
df_1['text'] = df_1['text'].apply(clean_text)
```

```python
df_1
```

```python
df_1[['text', 'label']].to_csv('cleaned_dataset_1.csv', index=False)
```
