```python
df['text'] = df['title'] + ' ' + df['body_text']
```

```python
import re

def clean_text(text):
    text = re.sub(r'<[^>]+>', '', text)  # ลบ HTML tags
    text = re.sub(r'[^\w\s]', '', text)  # ลบ special characters
    text = text.lower()  # แปลงเป็น lowercase
    return text
```

```python
df['preprocessed_text'] = df['text'].apply(clean_text)
df.head()
```

```python
# แปลงคอลัมน์วันที่เป็น datetime ก่อน
df['date'] = pd.to_datetime(df['date'])

date_ranges = (
    df.groupby('ticker')['date']
    .agg(start_date='min', end_date='max', record_count='count')
    .reset_index()
)

# คำนวณจำนวนวันจริง (รวมวันที่อาจไม่มีข้อมูล)
date_ranges['days_span'] = (date_ranges['end_date'] - date_ranges['start_date']).dt.days + 1

print(date_ranges[['ticker', 'start_date', 'end_date', 'days_span', 'record_count']])
```
